## Prerequisites

```python
import time
import os
import pandas as pd
import numpy as np
from tqdm import tqdm
import seaborn as sns
import matplotlib.pyplot as plt
import joblib
import warnings
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)
file_path = 'C:/Users/Koen.Janssen/OneDrive/Documents/TILBURG UNIVERSITY/Thesis/Data/'
parties = np.load('00_parties.npy')
```

## Performance

```python
dur = time.time()
results = pd.DataFrame(columns=['Party', 'Accuracy', 'Precision', 'Recall', 'F1 Score', 'Confusion Matrix'])
# -------------------------------------------------------------------------------------~----------------------------------------

for party in tqdm(parties, desc='Party loop'):                                         # for each party
    print(party)                                                                       # print party
    print("-----------------------------------------------------------------------")
    
    # load model
    # -------------------------------------------------------------------------
    model_url = file_path + f"50_models/{party}.pkl"                                   # define model url
    loaded_model = joblib.load(model_url)                                              # load model
    
    # load test data
    # -------------------------------------------------------------------------
    X_test_url = file_path + f"40_models_imbalanced/" + f"{party}_X_test.csv"          # define X_test url
    y_test_url = file_path + f"40_models_imbalanced/" + f"{party}_y_test.csv"          # define y_test url
    X_test = pd.read_csv(X_test_url)                                                   # read X_test
    X_test = X_test.drop(columns=['source', 'text','stemming_id','document_id'])       # redefine X_test, drop columns
    y_test = pd.read_csv(y_test_url)                                                   # read y_test
    
    # predict on test data
    # -------------------------------------------------------------------------
    y_pred = loaded_model.predict(X_test)                                              # predict y_test using X_test and model
    
    # calculate performance metrics
    # -------------------------------------------------------------------------
    accuracy = accuracy_score(y_test, y_pred)                                          # calculate performance metrics accuracy
    precision = precision_score(y_test, y_pred)                                        # calculate performance metrics precision
    recall = recall_score(y_test, y_pred)                                              # calculate performance metrics recall
    f1 = f1_score(y_test, y_pred)                                                      # calculate performance metrics f1
    cm = confusion_matrix(y_test, y_pred)                                              # calculate performance metrics cm
        
    # store results
    # -------------------------------------------------------------------------
    result = pd.DataFrame({
        'Party': [party],
        'Accuracy': [accuracy],
        'Precision': [precision],
        'Recall': [recall],
        'F1 Score': [f1],
        'Confusion Matrix': [cm]})
    results = pd.concat([results, result], ignore_index=True)                          # define result
    display(result)                                                                    # display result
    
    # display results
    # -------------------------------------------------------------------------
    sns.set(font_scale=1.2)                                                            # plot scale
    sns.set_style("whitegrid")                                                         # plot style
    fig, ax = plt.subplots(figsize=(8, 6))                                             # plot subplots
    sns.heatmap(cm, annot=True, fmt="g", cmap="Greys", cbar=False, annot_kws={"size": 14}, ax=ax)
    sns.despine()                                                                      # plot despine
    ax.set_xlabel("Predicted Label")                                                   # plot xlabel
    ax.set_ylabel("True Label")                                                        # plot ylabel
    ax.set_title(f"{party} Confusion Matrix")                                          # plot title
    plt.show()                                                                         # plot show
    print("\n\n\n")                                                                    # print spacers
    
# -------------------------------------------------------------------------------------~----------------------------------------
display(results)                                                                       # display results
print('\n---------------------------------------------------------------------------------------------------------------------')
print(f"Code duration: {round((time.time()  - dur),3)} seconds")    
```

```python
# Normalize the confusion matrix averages to represent ratios
average_metrics = results[['Accuracy', 'Precision', 'Recall', 'F1 Score']].mean().to_dict()
average_cm = np.mean(results['Confusion Matrix'].tolist(), axis=0)

# Normalize the confusion matrix ratio based on the total sum
total_sum = np.sum(average_cm)
average_cm_ratio = average_cm / total_sum

# Round the values in the confusion matrix ratio
average_cm_ratio_rounded = np.round(average_cm_ratio, 3)

# Create a DataFrame for averages
average_df = pd.DataFrame({
    'Party': ['Average'],
    'Accuracy': [average_metrics['Accuracy']],
    'Precision': [average_metrics['Precision']],
    'Recall': [average_metrics['Recall']],
    'F1 Score': [average_metrics['F1 Score']],
    'Confusion Matrix Ratio': [average_cm_ratio_rounded]
})

# Display the average DataFrame
display(average_df)
```

```python
sns.set(font_scale=1.2)
sns.set_style("whitegrid") 
fig, ax = plt.subplots(figsize=(8, 6))

# Explicitly set fmt='.2%'
sns.heatmap(average_cm_ratio, annot=True, fmt=".4%", cmap="Greys", cbar=False, annot_kws={"size": 14}, ax=ax)

sns.despine()
ax.set_xlabel("Predicted Label")
ax.set_ylabel("True Label")
ax.set_title("Average Confusion Matrix")
plt.show()
```
